Question 1 <br/> To perform a matrix multiplication between a matrix A of dimensions $ùëõ √ó ùëò$ and a matrix B of dimensions $k √ó m$, you need to calculate each element of the resulting matrix ùê∂ of dimensions $ùëõ √ó ùëö$.
<br/>
Each element $C_{ij}$ in the resulting matrix C is computed as:<br/>
$C_{ij}=\sum_{r=1}^{k}{A_{ir}.B_{rj}}$
For each element $C_{ij}$: <br/>
We need to perform k multiplications of $A_{ir}.B_{rj}$ and k-1 additions to add these products. 
<br/>
Since there are $m √ó n$ elements, total no. of multiplications are $m √ó n √ó k$ and total number of additions are $m √ó n √ó (k-1)$

```python
# Question 2
import numpy as np
import math
import time
def matrix_multiply_list(A, B):
    n = len(A)
    k = len(A[0])
    m = len(B[0])
    result = np.zeros((n, m))
    
    for i in range(n):
        for j in range(m):
            for l in range(k):
                result[i][j] += A[i][l] * B[l][j]
    return result
# Function to multiply two matrices using NumPy's dot function
def matrix_multiply_numpy(A, B):
    return np.dot(A, B)
```

```python
#Define matrix m1,m2
m1=np.random.randint(50, size=(4,6))
m2=np.random.randint(50, size=(6,10))
```

```python
# Timing the manual multiplication
start = time.time()
result_list = matrix_multiply_list(m1, m2)
end = time.time()
time_list = end - start
# Timing the NumPy multiplication
start = time.time()
result_np = matrix_multiply_numpy(m1, m2)
end = time.time()
time_np = end - start
print("List Multiply time",time_list,"\nNumpy multiply time",time_np)
if time_list>time_np:
    print("List Matrix Multiplication is faster")
else:
    print("Numpy matrix multiplication is faster")
```

```python
#Question 3
# Manual Bubble Sort to sort the array
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr
def median_bubble_sort(arr):
    sorted_arr = bubble_sort(arr)
    n = len(sorted_arr)
    if n % 2 == 0:
        return (sorted_arr[n // 2 - 1] + sorted_arr[n // 2]) / 2
    else:
        return sorted_arr[n // 2]
# time complexity is O(n^2)
```

```python
#A better method is mergesort
def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2
        left_half = arr[:mid]
        right_half = arr[mid:]
        
        merge_sort(left_half)
        merge_sort(right_half)
        
        i = j = k = 0
        
        while i < len(left_half) and j < len(right_half):
            if left_half[i] < right_half[j]:
                arr[k] = left_half[i]
                i += 1
            else:
                arr[k] = right_half[j]
                j += 1
            k += 1
        
        while i < len(left_half):
            arr[k] = left_half[i]
            i += 1
            k += 1
        
        while j < len(right_half):
            arr[k] = right_half[j]
            j += 1
            k += 1
    return arr
def median_merge_sort(arr):
    sorted_arr = merge_sort(arr)
    n = len(sorted_arr)
    if n % 2 == 0:
        return (sorted_arr[n // 2 - 1] + sorted_arr[n // 2]) / 2
    else:
        return sorted_arr[n // 2]
#time complexity here is O(nlogn)
```

```python
arr = np.random.randint(0, 100, size=1000)
```

```python
# Timing the Bubble Sort method
start = time.time()
median_bubble_sort_result = median_bubble_sort(arr)
end = time.time()
time_bubble_sort = end - start

# Timing the Merge Sort method
start = time.time()
median_merge_sort_result = median_merge_sort(arr)
end = time.time()
time_merge_sort = end - start

# Timing NumPy's median function
start = time.time()
median_numpy = np.median(arr)
end = time.time()
time_numpy = end - start

print(f"Time taken by Bubble Sort method: {time_bubble_sort} seconds")
print(f"Time taken by Merge Sort method: {time_merge_sort} seconds")
print(f"Time taken by NumPy median function: {time_numpy} seconds")

print(f"Median by Bubble Sort: {median_bubble_sort_result}")
print(f"Median by Merge Sort: {median_merge_sort_result}")
print(f"Median by NumPy: {median_numpy}")
```

```python
if time_bubble_sort < time_merge_sort and time_bubble_sort < time_numpy:
    print("Bubble sort is fastest")
elif time_merge_sort < time_bubble_sort and time_merge_sort < time_numpy:
    print("Merge sort is fastest")
elif time_numpy < time_bubble_sort and time_numpy<time_merge_sort:
    print("Numpy is fastest")
```

Question 4 <br/>The gradient is found out by the formula <br/> $\nabla f=\left(\frac{\partial f}{\partial x},\frac{\partial f}{\partial y}\right)$ <br/> Now $\frac{\partial f}{\partial x}$ = $2xy+y^3\cos{x}$ and $\frac{\partial f}{\partial y}$= $x^2+3y^2\sin{x}$ <br/> Therefore, gradient is $(2xy+y^3\cos{x},x^2+3y^2\sin{x})$

```python
#Question 6
import sympy as sp
x, y = sp.symbols('x y')
f = x**2 * y + y**3 * sp.sin(x)

# Compute the gradient
grad_f_x = sp.diff(f, x)  # Partial derivative with respect to x
grad_f_y = sp.diff(f, y)  # Partial derivative with respect to y

print(f"Partial derivative with respect to x: {grad_f_x}")
print(f"Partial derivative with respect to y: {grad_f_y}")

grad_f_x_val = grad_f_x.subs({x: x_val, y: y_val})
grad_f_y_val = grad_f_y.subs({x: x_val, y: y_val})

print(f"Gradient: ({grad_f_x_val},{grad_f_y_val})")
```

```python
#Question 8
class Database:
    class Year:
        def __init__(self, year):
            self.year = year
            self.branches = {}
        
        def add_branch(self, branch_name):
            self.branches[branch_name] = Database.Branch(branch_name)
        
        def __repr__(self):
            return f"Year(Year: {self.year}, Branches: {self.branches})"
    
    class Branch:
        def __init__(self, branch):
            self.branch = branch
            self.students = {}
        
        def add_student(self, rollno, name):
            self.students[rollno] = Database.Student(rollno, name)
        
        def __repr__(self):
            return f"Branch(Branch: {self.branch}, Students: {self.students})"
    
    class Student:
        def __init__(self, rollno, name):
            self.rollno = rollno
            self.name = name
            self.marks = {}
        
        def add_mark(self, subject, number):
            self.marks[subject] = Database.Marks(subject, number)
        
        def __repr__(self):
            return f"Student(Roll No: {self.rollno}, Name: {self.name}, Marks: {self.marks})"
    
    class Marks:
        def __init__(self, subject, number):
            self.subject = subject
            self.number = number
        
        def __repr__(self):
            return f"Marks(Subject: {self.subject}, Number: {self.number})"

    def __init__(self):
        self.years = {}

    def add_year(self, year):
        self.years[year.year] = year

    def __repr__(self):
        return f"Database(Years: {self.years})"
```

```python
#Question 13
def find_argmin(arr):
    min_index = 0 #index of min element
    for i in range(len(arr)):
        if arr[i] < arr[min_index]:
            min_index = i
    return min_index
    
```

```python
arr = np.random.randint(20, size=20)
print(arr)
```

```python
# Compute the index of the minimum element using customized function.
index_custom = find_argmin(arr)
print("Index of the minimum element (custom function):", index_custom)

# Compute the index of the minimum element using NumPy's argmin function for verification
index_np = np.argmin(arr)
print("Index of the minimum element (np.argmin):", index_np)
```
